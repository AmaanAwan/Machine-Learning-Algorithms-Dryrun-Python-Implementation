```
import numpy as np

X = np.array([[3, -1], [3, 1], [6,1], [6, -1], [1, 0], [0, 1], [-1, 0]])
y = np.array([1, 1, 1, 1, -1, -1, -1])  # Target Labels
```

```
learning_rate = 0.002
n_iters = 600

# Initializing parameters to 0
weights = np.zeros(X.shape[1])  # Number of features
bias = 0
```

cross-entropy loss / log loss formula:

```

# Gradient descent loop
for _ in range(n_iters):
  # Calculate linear combination (z)
  z = np.dot(X, weights) + bias

  # Sigmoid activation(A)
  A = 1 / (1 + np.exp(-z))

  # Binary cross-entropy loss / log loss
  epsilon = 1e-9
  loss = -np.mean(y * np.log(A + epsilon) + (1 - y) * np.log(1 - A + epsilon))

  # Calculate gradients: dz, dw, db
  dz = A - y
  print(_)
  print(loss)
  print(weights, bias)

  # Update weights and bias using gradient descent
  weights -= learning_rate * (1 / X.shape[0]) * np.dot(X.T, dz)
  bias -= learning_rate * (1 / X.shape[0]) * np.sum(dz)
```

```
# Prediction
threshold = 0.5
X_pred = np.array([0, -1])

y_hat = np.dot(X_pred, weights) + bias

y_predicted = 1 / (1 + np.exp(-y_hat))

y_predicted_cls = [1 if y_predicted > threshold else -1]

print("Predicted labels:", y_predicted_cls)
print("Loss after training:", loss)
```
