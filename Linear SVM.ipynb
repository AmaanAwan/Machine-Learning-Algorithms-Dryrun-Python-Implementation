# Linear Support Vector Machine

## Gradient Descent Approach

```
import numpy as np

X = np.array([[3, -1], [3, 1], [6,1], [6, -1], [1, 0], [0, 1], [-1, 0]]) 
y = np.array([1, 1, 1, 1, -1, -1, -1])  # Target Labels
```

 Learning rate controls the step size during weight updates 

the number of training epochs determines how many times the algorithm iterates through the data

weights(w) is initialized as a zero vector with the same dimension as the number of features in the data 

```
# Learning rate and number of epochs (iterations)
learning_rate = 0.01
epochs = 100

# Initialize weights (w) and bias (b) with zeros
w = np.zeros(X.shape[1])  # Weights have same dimension as features
b = 0
```

### Training

If condition(y[i] * y_pred <  1) is false ensures a minimum margin between the decision boundary and the correctly classified points.

If point is not classified correctly with margin, weights and bias are updated using the learning rate.

```

for epoch in range(epochs):

  # i represents each data point
  for i in range(len(X)):

    # Calculating predicted value
    y_pred = np.dot(X[i], w) + b

    
    if y[i] * y_pred < 1:  # Check for margin violation
      # Updating weights and bias
      w += learning_rate * (y[i] * X[i] - w)
      b += learning_rate * y[i]
```

Trained Weights and Bias

```
print("Weights:", w)
print("Bias:", b)
```

### Prediction

```

def predict(x):
  return np.dot(x, w) + b

# new data point
new_data = np.array([0, -1])
prediction = predict(new_data)


# prediction for the new data point
print("Prediction for", new_data, ":", prediction)

if prediction > 0:
    print("+1")
else:
    print("-1")
```
